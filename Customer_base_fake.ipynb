```python
import pandas as pd
from faker import Faker
from random import randint
```

```python
# Create instance of Faker
fake = Faker(
    locale = 'en-US'
)

# Set a seed to have fixed data each time
Faker.seed(1300)
```

```python
# Create a function to generate customer data where you can enter amount of unique customers to be created
def create_customers(num_employees):

    # Start with empty list
    customer_list = []

    # Create a dictionary for customers starting with 1 ending with entered amount
    for i in range(0, num_employees):
        customer = {}

        # Enter as much different list items as columns to describe each customer as needed
        customer['id'] = randint(3000,7999)
        customer['first_name'] = fake.first_name()
        customer['last_name'] = fake.last_name()
        customer['company'] = fake.company()
        customer['email'] = fake.email()
        customer['address'] = fake.address()
        customer['location'] = fake.local_latlng()
        customer['latitude'] = fake.latitude()
        customer['longtitude'] = fake.longitude()
        customer['state'] = fake.state()


        # Append created list items into the empty dictionary
        customer_list.append(customer)

    # Return a pandas dataframe as ending result of the function
    return pd.DataFrame(
        customer_list
    )
```

```python
create_customers(5)
```
